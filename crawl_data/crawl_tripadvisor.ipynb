```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import re 
```

```python
browser = webdriver.Edge(executable_path='driver/msedgedriver.exe')

# Maximize window
browser.maximize_window()

# Open google map
path_search = "https://www.tripadvisor.com.vn/Search?&q=Nh%C3%A0%20h%C3%A0ng&topIds=8113224%2C2397429%2C15336281%2C14938532%2C14135907&queryParsed=true&searchSessionId=63FBEE69CAF34E523EC182B68B73817C1660613129734ssid&sid=A7B1153EE3E91D811322E2AE8E32D0FD1660613198117&blockRedirect=true&ssrc=e"
browser.get(path_search)
sleep(3)
```

```python
try:
    show_more = browser.find_element(By.XPATH, "//span[@class='ui_link show-text']")
    show_more.click()
    sleep(1)
except:
    pass
```

```python
data = []
```

```python

while True:
    current_page_search = browser.current_url
    result_titles = browser.find_elements(By.XPATH, "//div[@class='result-title']")
    title_links = []
    for result_title in result_titles:
        # print(result_title)
        try:
            title_links.append(result_title.get_attribute("onclick").split(",")[3][2:-1])
        except:
            pass
    for title_link in title_links:
        browser.get(f"https://www.tripadvisor.com.vn{title_link}")
        sleep(5)
        while True:
            try:
                see_more = browser.find_element(By.XPATH, "//div[@class='see-more-mobile ui_button primary']")
                see_more.click()
                sleep(2)
            except:
                pass
            try:
                more_comment = browser.find_element(By.XPATH, "//span[@class='taLnk ulBlueLinks']")
                more_comment.click()
                sleep(2)
            except:
                pass
            comments = browser.find_elements(By.XPATH, "//div[@class='ui_column is-9']/div/div/p[@class='partial_entry']")
            for comment in comments:
                try:
                    data.append(comment.text)
                except:
                    pass
            next_comment_btn = None
            
            try:
                next_comment_btn = browser.find_element(By.XPATH, "//a[@class='nav next ui_button primary']")
            except:
                pass
            print(len(data))
            if next_comment_btn:
                next_comment_btn.click()
                sleep(5)
            else:
                break
    browser.get(current_page_search)
    sleep(5)
    try:
        show_more = browser.find_element(By.XPATH, "//span[@class='ui_link show-text']")
        show_more.click()
        sleep(1)
    except:
        pass
    next_page_btn = None
    try:
        next_page_btn = browser.find_element(By.XPATH, "//a[@class='ui_button nav next primary ']")
    except:
        pass
    if next_page_btn:
        next_page_btn.click()
        sleep(3)
    else:
        break


    
```

```python
data = pd.DataFrame({"comment": data})
```
